```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("HRDataset_v14.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

SCATTER PLT

```python
sns.scatterplot(data=df,x='Salary',y='EngagementSurvey',hue='Department',palette='rainbow')
```

LINE PLOT

```python
df['DateofHire']=pd.to_datetime(df['DateofHire'], errors='coerce')
```

```python
sns.lineplot(data=df,x='DateofHire',y='Salary',hue='Department',style='Department')
```

BAR PLOT

```python
sns.barplot(data=df, x='PerformanceScore',y='Salary',ci=None)
```

HISTOGRAM PLOT

```python
sns.histplot(data=df,x='Salary' ,hue='Department',kde=True)
```

KDE PLOT

```python
sns.kdeplot(data=df,x='Salary' ,hue='Department',fill=True,linewidth=1.5,)
```

BOX PLOT

```python
sns.boxplot(data=df,x='Department', y='Salary')
plt.xticks(rotation=30)   # Rotate x-axis labels
```

VIOLIN PLOT

```python
sns.violinplot(x='PerformanceScore', y='EngagementSurvey', data=df,hue='PerformanceScore',palette='dark')
```

HEAT MAP 

```python
cor=df.corr(numeric_only=True)
```

```python
cor
```

```python
plt.figure(figsize=(13,9)) 
sns.heatmap(cor,cmap='Blues',annot=True,)
```

PAIR PLOT

```python
sns.pairplot(data=df,hue='Department')
```

COUNT PLOT

```python
plt.xticks(rotation=20)
sns.countplot(x='Department', hue='Sex', data=df)
```
